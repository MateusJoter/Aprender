```python
import sqlite3 as sql
```

```python
# Conetando-se ao banco de dados
con = sql.connect(R'C:\Users\Mateus Joter\Desktop\Aprender\DATABASES\votacao.db')
cur = con.cursor()
```

```python
# Criação do Banco de Dados Teste:
# Moradores: nome e email funcionando como senha
# Votação 1 - Opções: opções que os moradores têm para votarem
# Votação 1 - Membro: tabela que vai conectar os moradores e as opções - ou seja, os votos

cur.executescript('''
CREATE TABLE IF NOT EXISTS moradores (
                            id INTEGER PRIMARY KEY AUTOINCREMENT,
                            nome TEXT UNIQUE NOT NULL,
                            email TEXT UNIQUE NOT NULL
                            );

CREATE TABLE IF NOT EXISTS votacao1_opcoes (
                            id INTEGER PRIMARY KEY AUTOINCREMENT,
                            opção TEXT UNIQUE NOT NULL
                            );
                            
CREATE TABLE IF NOT EXISTS votacao1_membro (
                            id_morador INTEGER NOT NULL,
                            id_voto INTEGER,
                            PRIMARY KEY (id_morador, id_voto)
                            )
                            ''')
```

```python
# Define e encerra

con.commit()
con.close()
print("Banco de Dados configurado!")
```
