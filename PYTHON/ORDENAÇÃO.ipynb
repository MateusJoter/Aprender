```python
nums = []

for i in ['primeiro', 'segundo', 'terceiro']:
    num = int(input(f'Digite o {i} número:'))
    nums.append(num)

crescente = True

for i in range(len(nums))[1:]:
    if nums[i] < nums[i-1]:
        crescente = False
        break

if crescente:
    print('crescente')
else:
    print('não está em ordem crescente')
```
